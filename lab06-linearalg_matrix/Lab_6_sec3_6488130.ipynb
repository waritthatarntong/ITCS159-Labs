**Exercise 1:** Consider the system of equations given below. Find a value for the coefficient $a$ so that the system has exactly one solution. For that value of $a$, find the solution to the system of equations and plot the lines formed by your equations to check your answer.

$$
\begin{eqnarray*}
ax_1 + 2x_2 & = & 1 \\
4x_1 + 8x_2 & = & 12 \\
\end{eqnarray*}
$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, solve, pprint

equ1, equ2, x1, x2, a = symbols('equ1, equ2, x1, x2, a')
equ1 = a*x1 + 2*x2 - 1
equ2 = 4*x1 + 8*x2 - 12
solutions = solve((equ1, equ2), (x1,x2))
pprint(solutions)
print("from this, we know that a can be anything except 1.\n")

print("For simplicity, I am going to assume that a = 0")
equ1 = equ1.subs(a, 0)
solutions = solve((equ1, equ2), (x1,x2))
pprint(f"if a = 0, then the solution of the system = ({solutions[x1]}, {solutions[x2]})")

# Plot the lines
x1 = np.linspace(-1, 4, 10)

fig, ax = plt.subplots()

# Assume a = 0
ax.plot(x1, x1 * 0 + 1/2)
ax.plot(x1, (12 - 4*x1)/8)

ax.text(0.1,0.3,'$x_2 = 1/2$')
ax.text(0.3,1.5,'$4x_1+8x_2 = 12$')

ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)
ax.grid(True, ls=':')
```

**Exercise 2:** Find a value for the coefficient $a$ so that the system has no solution. Plot the lines formed by your equations to verify there is no solution.

$$
\begin{eqnarray*}
 ax_1 + 2x_2 & = & 1 \\
4x_1 + 8x_2 & = & 12 \\
\end{eqnarray*}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, solve, pprint

equ1, equ2, x1, x2, a = symbols('equ1, equ2, x1, x2, a')
equ1 = a*x1 + 2*x2 - 1
equ2 = 4*x1 + 8*x2 - 12
solutions = solve((equ1, equ2), (x1,x2))
pprint(solutions)
print("from this, we know that when a = 1 the system has no solutions")
      
equ1 = equ1.subs(a, 1)
solutions = solve((equ1, equ2), (x1,x2))
pprint(f"The system has no solutions: {solutions}")

# Plot the lines
x1 = np.linspace(-1, 4, 10)

fig, ax = plt.subplots()

# Assume a = 1
ax.plot(x1, (1-x1)/2)
ax.plot(x1, (12 - 4*x1)/8)

ax.text(0.1,-0.3,'$x_1+2x_2 = 1$')
ax.text(0.3,1.5,'$4x_1+8x_2 = 12$')

ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)
ax.grid(True, ls=':')

print("The graphs are parallel, therefore the system has no solutions")
```

**Exercise 4:**  Solve the system $AX=B$ by finding $A^{-1}$ and computing $X=A^{-1}B$.

$$
A = \left[ \begin{array}{rrrr} 1 & 2 & -3 \\ -2 & 2 & -2  \\ 0 & -2 & 3  \end{array}\right] \quad\quad
B = \left[ \begin{array}{rrrr} 1  \\ 2 \\ 1  \end{array}\right]
$$    

```python
import numpy as np

A, B, Ainv = symbols('A, B, Ainv')
A = np.array([[1,2,-3], [-2,2,-2], [0,-2,3]])
B = np.array([[1,2,1]]).transpose()

# Find A inverse
Ainv = np.linalg.inv(A)
# print("Ainv x A = \n", np.round(A@Ainv))

# Compute X = Ainv x B
X = Ainv @ B
print("Ainv x B =")
print(X)
```

**Exercise 5**:
Hill cipher is a polygraphic substitution cipher based on linear algebra.Each letter is represented by a number modulo 26. Often the simple scheme A = 0, B = 1, …, Z = 25 is used, but this is not an essential feature of the cipher. To encrypt a message, each block of n letters (considered as an n-component vector) is multiplied by an invertible n × n matrix, against modulus 26. To decrypt the message, each block is multiplied by the inverse of the matrix used for encryption.
The matrix used for encryption is the cipher key, and it should be chosen randomly from the set of invertible n × n matrices (modulo 26).

**Examples**: 

- Input  : Plaintext: ACT
- Key: GYBNQKURP
- Output : Ciphertext: POH

- Input  : Plaintext: GFG
- Key: HILLMAGIC 
- Output : Ciphertext: SWK

**Encryption**

We have to encrypt the message **ACT (n=3)**.
The **key is GYBNQKURP** which can be written as the nxn matrix : 
$$
\begin{equation}
B = \left[ \begin{array}{ccc} 6 & 24 & 1  \\  13 & 16 & 10 \\  20 & 17 & 15\\  \end{array}\right]
\end{equation}
$$  

The message 'ACT' is written as vector: 
$$
\begin{equation}
P = \left[ \begin{array}{c} 0  \\  2 \\  19 \\  \end{array}\right]
\end{equation}
$$

The enciphered vector is given as: 
$$
C = B*P
$$
$$
\begin{equation}
C = \left[ \begin{array}{ccc} 6 & 24 & 1  \\  13 & 16 & 10 \\  20 & 17 & 15\\  \end{array}\right] 
\left[ \begin{array}{c} 0  \\  2 \\  19 \\  \end{array}\right]
= 
\left[ \begin{array}{c} 67  \\  222 \\  319 \\  \end{array}\right]
\end{equation}
=
\left[ \begin{array}{c} 15  \\  14 \\  7 \\  \end{array}\right]
\left( mod 26 \right)
$$  

which corresponds to ciphertext of **POH** 


**Decryption**

To decrypt the message, we turn the ciphertext back into a vector, then simply multiply by the inverse matrix of the key matrix (IFKVIVVMI in letters).

The **inverse of the matrix** used in the previous example is: 
$$
   B^{-1} = 
    \left[ \begin{array}{ccc} 6 & 24 & 1  \\  13 & 16 & 10 \\  20 & 17 & 15\\  \end{array}\right]^{-1}
    =
    \left[ \begin{array}{ccc} 8 & 5 & 10  \\  21 & 8 & 21 \\  21 & 12 & 8\\  \end{array}\right] 
    \left( mod 26 \right)
$$

For Ciphertext **POH**, the decript text is

$$
   D = 
     \left[ \begin{array}{ccc} 8 & 5 & 10  \\  21 & 8 & 21 \\  21 & 12 & 8\\  \end{array}\right] 
    \left[ \begin{array}{c} 15  \\  14 \\  7 \\  \end{array}\right]
   = \left[ \begin{array}{c} 260  \\  574 \\  739 \\  \end{array}\right]
   = \left[ \begin{array}{c} 0  \\  2 \\  19 \\  \end{array}\right]
   \left( mod 26 \right)
 $$
which gives us back ‘ACT’. 
Assume that all the alphabets are in upper case.

```python
import numpy as np
from sympy import Matrix

# Encrypting: C = BP
print("Encrypting")
# Key
print("The key is GYBNQKURP")
B = np.array([[6,24,1],[13,16,10],[20,17,15]])

# The message 'ANT'
print("The message is 'ANT' = ")
P = np.array([[0,13,19]]).transpose()
print(P)

# Ciphertext
C = (B@P) % 26
print("\nThe ciphertext is 'TIM' =")
print(C)

# Decrypting: P = BinvC
print("\nDecrypting 'TIM'")
B = Matrix([[6,24,1],[13,16,10],[20,17,15]])
Binv = B.inv_mod(26)
_P = (Binv@C) % 26
print(f"Result: {_P}\n")
print("'TIM' can be decrypted into 'ANT'.")
```
