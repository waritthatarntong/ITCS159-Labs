1. For any principal amount $p$, any rate $r$, and any number of years $t$, the 
compound interest is calculated using the formula:
$$ A = P\left(1+\frac{r}{n}\right)^{nt}$$ 

Assume you are 20 year old now, you deposited 1 Million bath to bank this year. If you wait until retirement, interest paid __Semiannually__ for 8%.

How much AMOUNT in million bath you could widthdraw at after 60 year old?

```python
from sympy import symbols, evalf, plot
P, n, r, t,x= symbols('P,n,r,t,x')

A = P*(1 + r/n)**(n*t)

# Independent variable (parameter) = variable that change = t
print(A.subs({P:1e6, n:2, r:0.08, t:40}))

# Dependent variable = P, n, r
A = A.subs({P:1e6, n:2, r:0.08})
plot(A, (t, 1, 40))
```

2.  Given $ y= -x^2 + 2x +15 $ plot graph of y(x), y'(x) and solve for maximum value of y(x). 

```python
from sympy import symbols, plot, diff, solve

x = symbols('x')
y = -x**2 + 2*x + 15
dy = diff(y)

plot(y, dy)

# Critical point
critical_pt = solve(dy)
print(f'Critical point is at {critical_pt[0]}')
print('And because graph y is a downward parabola, we know that the critical point is also a maximum point.')
print(f'Maximum value of y(x) = {critical_pt[0]}')
```

3. Given $ y= -x^2 + 2x + 15 $  compute intersection points,compute area below curve y(x) above y(x)= 15. 

```python
from sympy import symbols, plot, integrate, solve

x = symbols('x')
y = -x**2 + 2*x + 15
plot(y, 15, (x, -3, 6))

# when y = 15
# -x**2 + 2x 
intersection_pts = solve(y - 15)
print(f'Intersection points: {intersection_pts}')

area_y = integrate(y, (x, intersection_pts[0], intersection_pts[1]))
rectangle = 15 * (intersection_pts[1] - intersection_pts[0])

area_above_y = area_y - rectangle
print(f'Area above y: {area_above_y}')
```
